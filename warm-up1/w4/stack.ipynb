# Stack
- Last In First Out (LIFO)
- Pre-defined capacity (Limited in size)



- **Push**: add an element to the top of a stack



- **Pop**: remove an element from the top of a stack



- **Overflow**: try to push an element to a full stack



- `is_full`: Check if the stack is full

- **Underflow**: try to pop out an element to an empty stack



- `is_empty`: Check if the stack is empty



```python
# Implement Stack
class MyStack:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__stack = []
    
    def is_empty(self):
        return len(self.__stack) == 0
    
    def is_full(self):
        return len(self.__stack) == self.__capacity
    
    def pop(self):
        if self.is_empty():
            raise Exception("Underflow")
        else:
            return self.__stack.pop()
    
    def push(self, value):
        if self.is_full():
            raise Exception("Overflow")
        else:
            return self.__stack.append(value)
        
    def top(self):
        if self.is_empty():
            print("Queue is empty!")
            return
        else:
            return self.__stack[-1]

    
```

```python
stack = MyStack(2)
stack.push(3)  
stack.top()
```
