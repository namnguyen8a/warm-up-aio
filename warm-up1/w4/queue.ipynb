# Queue
- First In First Out (FIFO)
- Pre-defined capacity (Limited size)



- **Enqueue**: Add an element to the end of the queue



- **Dequeue**: Remove an element from the front of the queue



- **Overflow**: Try to enqueue an element to a full queue



- `is_full`: Check if the queue is full
- **Underflow**: Try to dequeue an empty queue



- `is_empty`: Check if the queue is empty

```python
class MyQueue:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__queue = []
    
    def is_empty(self):
        return len(self.__queue) == 0

    def is_full(self):
        return len(self.__queue) == self.__capacity

    def dequeue(self):
        if self.is_empty():
            raise Exception("Underflow")
        return self.__queue.pop(0) # pop(0) remove 1st ele , pop(1) remove 2nd,... from left to right ->
    
    def enqueue(self, value):
        if self.is_full():
            raise Exception("Overflow")
        return self.__queue.append(value)
    def front(self):
        if self.is_empty():
            print("Queue is empty!")
            return
        else:
            return self.__queue[0]
```

```python
q = MyQueue(4)
q.enqueue(2)
q.front()
```
