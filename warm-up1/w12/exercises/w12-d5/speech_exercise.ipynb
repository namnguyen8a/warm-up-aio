```python
test_case_1 = "data/tt_1.wav"   # output: các đồ thị và đoạn văn bản tương ứng với file truyền vào
test_case_2 = "Tôi yêu AIO"     # output: một đoạn âm thanh mp3 với nội dung là "Tôi Yêu AIO"

# Dồ thị tương ừng với file test_case_1
import librosa
audio, sr = librosa.load(test_case_1)

import numpy as np
import matplotlib.pyplot as plt
time = np.linspace(0, audio.shape[0] / sr, num=audio.shape[0])
plt.plot(time, audio, color="blue")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (quantized)")
plt.title("Wav file visualization")
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
# Một đoạn văn bản tương ừng với test_case_1
import speech_recognition as sr
r = sr.Recognizer()

my_audio = sr.AudioFile(test_case_1)
with my_audio as source:
    audio = r.record(source)

print(type(audio))
your_speech = r.recognize_google(audio, language="vi-VN")
print("Audio transcription: ", your_speech)
```

```python
# Một đoạn âm thanh mp3 với nội dung là "Tôi yêu AIO"
from gtts import gTTS
import librosa
import numpy as np

lang='vi'
ouput_filename = 'data/audio1.mp3'
output = gTTS(test_case_2, lang=lang, slow=False)

output.save(ouput_filename)
data, sr = librosa.load(ouput_filename)

import IPython
IPython.display.display(IPython.display.Audio(np.transpose(data), rate=sr))
```
