# Sử dụng thư viện librosa để visualize âm thanh

```python
import librosa

filename = librosa.example('nutcracker')
# Load the audio as a waveform 'y'
# Store the sampling rate as 'sr'
audio, sr = librosa.load(filename)

import numpy as np
import matplotlib.pyplot as plt
time = np.linspace(0, audio.shape[0] / sr, num=audio.shape[0])
plt.plot(time, audio, color="blue")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (quantized)")
plt.title("Wav file visualization")
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

# Sử dụng thư viện SpeechRecognition để chuyển âm thanh thành văn bản

```python
import speech_recognition as sr

r = sr.Recognizer()
audio_filename = 'data/tt_1.wav'

my_audio = sr.AudioFile(audio_filename)
with my_audio as source:
    audio = r.record(source)

print(type(audio))
your_speech = r.recognize_google(audio, language="vi-VN")
print("Audio transcription: ", your_speech)
```

# gtts cho phép tạo và phát lại giọng nói từ văn bản sử dụng TTS của google

```python
%pip install ipython
```

```python
from gtts import gTTS
import librosa
import numpy as np

lang='vi'
output_filename = 'data/record.mp3'
content = "Xin chào mọi người"
output = gTTS(content, lang=lang, slow=False) # text to speech

output.save(output_filename) # save goole audio to a file
data, sr = librosa.load(output_filename) # load google audio

import IPython
IPython.display.display(IPython.display.Audio(np.transpose(data), rate=sr))
```
